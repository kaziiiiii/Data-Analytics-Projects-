```python
# Create 2 new columns 
## mins per match 
# goals per match 

epl_data['MinsPerMatch'] = (epl_data['Mins'] / epl_data['Matches']).astype(int)
epl_data['GoalsPerMatch'] = (epl_data['Goals'] / epl_data['Matches']).astype(float)
epl_data.head()
```

```python
# Total Goals   ===> Total Goals score in 2020-2021 epl 

Total_Goals = epl_data['Goals'].sum()
print(Total_Goals)
```

```python
# Penalty Goals

Total_PenaltyGoals = epl_data['Penalty_Goals'].sum()
print(Total_PenaltyGoals)
```

```python
# Penalty Attempts 

Total_PenaltyAttempts = epl_data['Penalty_Attempted'].sum()
print(Total_PenaltyAttempts)
```

```python
# Pie chart for penalties  scored  vs missed

plt.figure(figsize=(20,10))
pl_not_scored = epl_data['Penalty_Attempted'].sum() - Total_PenaltyGoals
data = [pl_not_scored,Total_PenaltyGoals]
lables = ['Penalties missed' , 'Penalties Scored']
color = sns.color_palette("Set2")
plt.pie(data, labels = lables, colors = color, autopct = '%.0f%%')
plt.show()

```
